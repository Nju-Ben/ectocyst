```python
#load data
import pandas as pd
#data = pandas.read_csv("Stream.csv",encoding)
data = pd.read_excel(r'Stream.xlsx')
data = data[0:500]
data.head(5)
```

```python
#import package
import re
from nltk.corpus import stopwords 
from nltk.stem import WordNetLemmatizer
```

```python
def my_filter(str):
    """
        filtering text
        :param text: the string of text
        :return: text string after filtering
    """
    
    #remove uerls
    newS=''
    url_reg  = r'[a-z]*[:.]+\S+'
    for s in str:
        if s.isalpha() or s=='@' or s in '/:.\' ':
            newS += s
    newS   = re.sub(url_reg, '', newS)
    
    #remove punctuation
    newSS=''
    for s in newS:
        if s.isalpha() or s==' ':
            newSS += s
    return newSS
```

```python
# change letter to lower
def change2lower(str1):
    str2=str1.lower()
    return str2
```

```python
def stemmed(str1):
    temp=str1.split(' ')
    
    #temp=[w for w in words.words()]
    #stem=[PorterStemmer().stem(w) for w in temp]
    
    #lemmatize words
    stem=[WordNetLemmatizer().lemmatize(w) for w in temp]
    
    #remove stopwords
    stem = [w for w in stem if w not in stopwords.words('english')]
    
    #remove unrelated word rt
    if 'rt' in stem:
        stem.remove('rt')
        
    # concact word list to str
    result=' '.join(stem)
    return result
```

```python
def clean_text(text):
        """
        Clean text
        :param text: the string of text
        :return: text string after cleaning
        """
        # acronym
        text = re.sub(r"can\'t", "can not", text)
        text = re.sub(r"cannot", "can not ", text)
        text = re.sub(r"what\'s", "what is", text)
        text = re.sub(r"What\'s", "what is", text)
        text = re.sub(r"\'ve ", " have ", text)
        text = re.sub(r"n\'t", " not ", text)
        text = re.sub(r"i\'m", "i am ", text)
        text = re.sub(r"I\'m", "i am ", text)
        text = re.sub(r"\'re", " are ", text)
        text = re.sub(r"\'d", " would ", text)
        text = re.sub(r"\'ll", " will ", text)
        text = re.sub(r" e mail ", " email ", text)
        text = re.sub(r" e \- mail ", " email ", text)
        text = re.sub(r" e\-mail ", " email ", text)

        # spelling correction
        text = re.sub(r"ph\.d", "phd", text)
        text = re.sub(r"PhD", "phd", text)
        text = re.sub(r" e g ", " eg ", text)
        text = re.sub(r" fb ", " facebook ", text)
        text = re.sub(r"facebooks", " facebook ", text)
        text = re.sub(r"facebooking", " facebook ", text)
        text = re.sub(r" usa ", " america ", text)
        text = re.sub(r" us ", " america ", text)
        text = re.sub(r" u s ", " america ", text)
        text = re.sub(r" U\.S\. ", " america ", text)
        text = re.sub(r" US ", " america ", text)
        text = re.sub(r" American ", " america ", text)
        text = re.sub(r" America ", " america ", text)
        text = re.sub(r" mbp ", " macbook-pro ", text)
        text = re.sub(r" mac ", " macbook ", text)
        text = re.sub(r"macbook pro", "macbook-pro", text)
        text = re.sub(r"macbook-pros", "macbook-pro", text)
        text = re.sub(r" 1 ", " one ", text)
        text = re.sub(r" 2 ", " two ", text)
        text = re.sub(r" 3 ", " three ", text)
        text = re.sub(r" 4 ", " four ", text)
        text = re.sub(r" 5 ", " five ", text)
        text = re.sub(r" 6 ", " six ", text)
        text = re.sub(r" 7 ", " seven ", text)
        text = re.sub(r" 8 ", " eight ", text)
        text = re.sub(r" 9 ", " nine ", text)
        text = re.sub(r"googling", " google ", text)
        text = re.sub(r"googled", " google ", text)
        text = re.sub(r"googleable", " google ", text)
        text = re.sub(r"googles", " google ", text)
        text = re.sub(r"dollars", " dollar ", text)
        
        return text
```

```python
data["Tweet content new"]=data["Tweet content"].apply(clean_text)
data["Tweet content new"]=data["Tweet content new"].apply(my_filter)
data['content_afterfilter']=data["Tweet content new"].apply(change2lower)
data['content_final']=data['content_afterfilter'].apply(stemmed)
data.head(5)
```

```python
#import package for sentiment analysis
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()
```

```python
def cal_sentiment(str1):
    vs = analyzer.polarity_scores(str1)
    score=vs['compound']
    if score >0.05:
        return 1
    elif score <= 0.05 and score >= -0.05:
        return 0
    else:
        return -1
"""
The compound score is computed by summing the valence scores of each word in the lexicon, adjusted according to the rules, 

and then normalized to be between -1 (most extreme negative) and +1 (most extreme positive). This is the most useful metric 

if you want a single unidimensional measure of sentiment for a given sentence. Calling it a 'normalized, weighted composite 

score' is accurate.


It is also useful for researchers who would like to set standardized thresholds for classifying sentences as either positive,

neutral, or negative. Typical threshold values (used in the literature cited on this page) are:

positive sentiment: compound score >= 0.05

neutral sentiment: (compound score > -0.05) and (compound score < 0.05)

negative sentiment: compound score <= -0.05


"""
```

```python
data['sentiment']=data['content_final'].apply(cal_sentiment)
final_result=data[['Tweet content','content_final','sentiment']]
final_result
```
